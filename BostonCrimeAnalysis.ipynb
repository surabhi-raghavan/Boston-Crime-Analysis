```python
# Map the final 7 offense codes
final_mapping = {
    650: 'Part One',     # Migrated Report - Larceny From MV
    1200: 'Part Two',    # Migrated Report - Embezzlement
    1400: 'Part Two',    # Migrated Report - Vandalism/Destruction of Property
    2400: 'Part Two',    # Migrated Report - Affray/Disturbing the Peace
    2500: 'Part Two',    # Migrated Report - Kidnapping
    2950: 'Part Three',  # Migrated Report - Motor Vehicle Crash
    3350: 'Part Three'   # Migrated Report - Injured/Medical/Sick Assist
}

# Apply the final mapping
combined_df['UCR_PART'] = combined_df['UCR_PART'].fillna(combined_df['OFFENSE_CODE'].map(final_mapping))

# Verify that there are no more missing values in UCR_PART
print(f"Final missing UCR_PART values: {combined_df['UCR_PART'].isna().sum()}")
```

```python
# Define mapping for the remaining unmapped offenses based on UCR guidelines
remaining_mapping = {
    530: 'Part One',    # BREAKING AND ENTERING (B&E) MOTOR VEHICLE (Property Crime)
    600: 'Part One',    # Migrated Report - Other Larceny (Property Crime)
    400: 'Part One',    # Migrated Report - Aggravated Assault (Violent Crime)
    100: 'Part One',    # Migrated Report - Criminal Homicide (Violent Crime)
    500: 'Part One',    # Migrated Report - Burglary/Breaking and Entering (Property Crime)
    700: 'Part One',    # Migrated Report - Auto Theft (Property Crime)
    300: 'Part One',    # Migrated Report - Robbery (Violent Crime)

    121: 'Part Two',    # MANSLAUGHTER - VEHICLE - NEGLIGENCE
    122: 'Part Two',    # MANSLAUGHTER - NEGLIGENCE
    800: 'Part Two',    # Migrated Report - Assault/Assault & Battery (Simple Assault)
    1800: 'Part Two',   # Migrated Report - Drugs - Possession/Manufacture
    1100: 'Part Two',   # Migrated Report - Fraud
    1000: 'Part Two',   # Migrated Report - Counterfeiting/Forgery
    1500: 'Part Two',   # Migrated Report - Weapons Violation
    2600: 'Part Two',   # Migrated Report - Other Part II

    3000: 'Part Three', # Migrated Report - Death Investigation
    3300: 'Part Three', # Migrated Report - Other Part III
    3200: 'Part Three', # Migrated Report - Investigate Person
    3100: 'Part Three', # Migrated Report - Investigate Property

    99999: 'Other'      # Evidence Tracker Incidents
}

# Apply the mapping to fill any of these specific missing values
combined_df['UCR_PART'] = combined_df['UCR_PART'].fillna(combined_df['OFFENSE_CODE'].map(remaining_mapping))

# Check how many missing values remain now
final_remaining_missing = combined_df['UCR_PART'].isna().sum()
print(f"Missing UCR_PART values after final mapping: {final_remaining_missing}")

# Display any remaining edge cases if they exist
if final_remaining_missing > 0:
    display(combined_df[combined_df['UCR_PART'].isna()][['OFFENSE_CODE', 'OFFENSE_DESCRIPTION']].drop_duplicates().head(20))
```

```python
# Create a dictionary mapping from OFFENSE_CODE to UCR_PART based on existing valid entries
ucr_mapping = combined_df.dropna(subset=['UCR_PART']).set_index('OFFENSE_CODE')['UCR_PART'].to_dict()

# Fill the missing UCR_PART values using this mapping
combined_df['UCR_PART'] = combined_df['UCR_PART'].fillna(combined_df['OFFENSE_CODE'].map(ucr_mapping))

# Check how many missing values remain
remaining_missing = combined_df['UCR_PART'].isna().sum()
print(f"Missing UCR_PART values after mapping: {remaining_missing}")

# If any remain, let's look at the top unmapped offense descriptions to classify them
if remaining_missing > 0:
    print("\nTop unmapped offenses:")
    unmapped_offenses = combined_df[combined_df['UCR_PART'].isna()][['OFFENSE_CODE', 'OFFENSE_DESCRIPTION']].drop_duplicates()
    display(unmapped_offenses.head(20))
```
